# Linked List

### Linked List vs List
- Linked list doesnt have an index
- List is are store in a contiguous place in memory.
- Linked List are not concurrent in memomory, Each Nodes point to the address of the following node
- Linked list will have a head and tail attribute to ppoint to end and tail 

### Big O

_Appended_
- O(1) Pointer at tail now points to new number.

_Removing from end of link list_
- O(n) Traverse through list to get node pointing last number.

_Adding to the front_
- O(1) Pointer for header now points to the new number.

_Removing header_
- 0(1) Pointer at first number now points to second

_Adding randomly__
- 0(n) start and iterate througout the list. Set new item pointer equal to old item, old item point to new

_Look Up By Value or Index_
- 0(n) You have to start at the head and iterate to find anything



```python
class Node:
    def __init__(self, value: int):
        self.value: int = value
        self.next: Node = None
```

```python
class LinkedList:
    def __init__(self,value: int):
        new_node: Node = Node(value)
        self.head: Node = new_node
        self.tail: Node = new_node
        self.length: int = 1
    
    def print_list(self):
        x: Node = self.head
        while x is not None and self.length != 0:
            print(x.value)
            x: Node = x.next

    def append(self, value: int):
        new_node = Node(value)
        if self.length != 0:
            self.tail.next: Node = new_node
            self.tail: Node = new_node
        else:
            self.tail: Node = new_node
            self.head: Node = new_node
        self.length += 1
    
    def pop(self, value: int):
        current_node: Node = self.head
        next_node: Node = current_node.next
            
        while next_node.value != value and next_node != None:
            current_node: Node = current_node.next
            next_node: Node = current_node.next
        
        left_node: Node = current_node
        right_node: Node = left_node.next
        left_node.next: Node = right_node.next
 
    def prepend(self, value: int):
        new_head: Node = Node(value)
        new_head.next: Node = self.head.next
        self.head: Node = new_head
        
    def pop_first(self):
        if self.length != 0:
             self.head: Node = self.head.next
        else:
            print("Chill out...")
        
    def insert(self, value, position):
        pass
    def remove(self):
        pass
    def lookup_by_index(self):
        pass
    def lookup_by_value(self):
        pass
```
